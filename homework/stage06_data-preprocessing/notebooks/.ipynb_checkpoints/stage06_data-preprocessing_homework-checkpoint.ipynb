```python
import pandas as pd
import os
import sys 
from src.cleaning import fill_missing_median, drop_missing, normalize_data

raw_df = pd.read_csv("../data/raw/sample_data.csv")
raw_df.head()
```

```python
df = raw_df.copy()

# Fill missing
df = fill_missing_median(df, "Volume")

# Drop columns with too much missingness
df = drop_missing(df, threshold=0.7)

# Normalize numeric cols
df = normalize_data(df, ["Open", "High", "Low", "Close"])
```

```python
print("Before cleaning:", raw_df.shape)
print("After cleaning:", df.shape)

raw_df.describe()
df.describe()
```

```python
df.to_csv("../data/processed/cleaned_dataset.csv", index=False)
df.to_parquet("../data/processed/cleaned_dataset.parquet", engine="fastparquet", index=False)
```
